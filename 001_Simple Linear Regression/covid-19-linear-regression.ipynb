```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
covid_india = pd.read_csv("covid_19_india.csv")
covid_india
```

```python
print covid_india.isnull().sum()
print covid_india.shape
```

```python
covid_india.dropna(inplace=True)
print covid_india.isnull().sum()
print covid_india.shape
```

```python
No_of_days = covid_india.iloc[:,0]
confr = covid_india.iloc[:,-1]
print No_of_days
print "*************"
print confr
```

```python
print No_of_days.shape
print confr.shape
print No_of_days.ndim
print confr.ndim
```

```python
X = No_of_days[:,np.newaxis]
y = confr[:,np.newaxis]
print X.shape
print y.shape
print X.ndim
print y.ndim
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8)
print X_train.shape
print y_train.shape
print X_test.shape
print y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
m= lr.coef_
c = lr.intercept_
print "confirmed = ",m," *  Days + ",c
```

```python
y_prd  =lr.predict(X_test)
plt.scatter(X_test,y_test)
plt.plot(X_test,y_prd,c="r")
plt.show()
```

```python
con = np.arange(len(y_test))[:,np.newaxis]
plt.plot(con,y_test,"g")
plt.plot(con,y_prd,"r")
plt.title("showing original data & predicted data")
plt.ylabel("covid cases")
plt.xlabel("Constant")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print mean_squared_error(y_test,y_prd)
print r2_score(y_test,y_prd)
```

```python
con = np.arange(len(y_test))[:,np.newaxis]
plt.plot(con,y_test- y_prd,"g")
plt.title("Error Term")
plt.ylabel("covid cases")
plt.xlabel("Constant")
plt.show()
```
